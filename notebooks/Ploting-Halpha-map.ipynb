```python
import numpy as np
from astropy.io import fits
from scipy.ndimage import rotate
import matplotlib.pyplot as plt
from matplotlib.transforms import Affine2D
import aplpy
from matplotlib.patches import Rectangle
from astropy.visualization import (MinMaxInterval, AsinhStretch)
from astropy.wcs import WCS
from astropy.visualization import wcsaxes
from astropy import units as u
import pandas as pd
from astropy.table import Table
import os
```

```python
def plot_halpha_map(file_path, output_path):
    try:
        # Open the FITS file
        hdul = fits.open(file_path)
        hdu = hdul[0]  # Get the primary HDU

        # Calculate mean and standard deviation for color scaling
        m, s = np.mean(hdu.data), np.std(hdu.data)

        # Create a new figure
        fig = plt.figure(figsize=(8, 8))

        # Create APLpy FITSFigure instance
        f = aplpy.FITSFigure(hdu, figure=fig,  subplot=(1, 1, 1), north=True)

        # Show the image with a specific colormap and color bar
        f.show_colorscale(cmap='viridis', vmin=m, vmax=m + 2 * s)  # Adjust the intensity range

        # Add a color bar
        f.add_colorbar()
        f.colorbar.set_location("right")
        f.colorbar.set_axis_label_text(r"F(H$\alpha$ + [NII])")
        f.colorbar.set_font(size='x-large',  weight='medium', \
                      stretch='normal', family='sans-serif', \
                      style='normal', variant='normal')
        f.colorbar.set_axis_label_font(size=18)#, weight='bold')

        # Set axis labels to display in WCS coordinates
        f.axis_labels.set_xtext('RA (J2000)')
        f.axis_labels.set_ytext('DEC (J2000)')
        f.axis_labels.set_font(size=18, weight='medium', stretch='normal', family='sans-serif', style='normal', variant='normal')
        f.tick_labels.set_font(size=18, weight='medium', stretch='normal', family='sans-serif', style='normal', variant='normal')

        # Hide axis and tick labels
        #f.axis_labels.hide()
        #f.tick_labels.hide()
        f.add_scalebar(20/3600.) #20
        f.scalebar.set_label('20"')
        f.scalebar.set(color='white', linewidth=4, alpha=0.9)
        f.scalebar.set_font(size=45, weight='bold',
                      stretch='normal', family='sans-serif',
                      style='normal', variant='normal')
        f.frame.set_linewidth(1)  # points
        f.frame.set_color('black')

        # Save the plot as a PDF file
        plt.savefig(output_path)
        plt.close(fig)

        hdul.close()  # Close the FITS file after use

    except FileNotFoundError:
        print(f"File {file_path} not found.")
    except Exception as e:
        print(f"Error plotting {file_path}: {e}")
```

```python
# Example usage
directory = "../maps_fits/"  # Directory containing FITS files
```

### Plotting

```python
# Iterate through files in the input directory
for file_name in os.listdir(directory):
    if file_name.endswith(".fits"):  # Process only FITS files
        file_path = os.path.join(directory, file_name)
        output_path = os.path.join(directory, file_name.replace(".fits", ".pdf"))

        plot_halpha_map(file_path, output_path)
```
