```python
# Remove single quotes
df['ID'] = df['ID'].str.replace("b'", '')
df['ID'] = df['ID'].str.replace("'", '')
df
```

```python
Query = f"""SELECT r.Field, r.ID, r.RA_r, r.DEC_r, r.X_r, r.Y_r, r.s2n_r_psf, j660.s2n_J0660_psf,
            i.s2n_i_psf, r.CLASS_STAR_r, i.CLASS_STAR_i, u.u_psf, j378.J0378_psf, 
            j395.J0395_psf, j410.J0410_psf, j430.J0430_psf, g.g_psf, j515.J0515_psf, 
            r.r_psf, j660.J0660_psf, i.i_psf, j861.J0861_psf, z.z_psf, 
            u.e_u_psf, j378.e_J0378_psf, j395.e_J0395_psf, j410.e_J0410_psf, j430.e_J0430_psf, 
            g.e_g_psf, j515.e_J0515_psf, r.e_r_psf, j660.e_J0660_psf, i.e_i_psf, j861.e_J0861_psf, z.e_z_psf 
            FROM TAP_UPLOAD.upload as tap LEFT JOIN "idr4_psf"."idr4_psf_u" as u ON tap.ID= u.ID  
            LEFT JOIN "idr4_psf"."idr4_psf_j0378" as j378 ON tap.ID=j378.ID 
            LEFT JOIN "idr4_psf"."idr4_psf_j0395" as j395 ON tap.ID=j395.ID
            LEFT JOIN "idr4_psf"."idr4_psf_j0410" as j410 ON tap.ID=j410.ID 
            LEFT JOIN "idr4_psf"."idr4_psf_j0430" as j430 ON tap.ID=j430.ID 
            LEFT JOIN "idr4_psf"."idr4_psf_g" as g ON tap.ID=g.ID 
            LEFT JOIN "idr4_psf"."idr4_psf_j0515" as j515 ON tap.ID=j515.ID 
            LEFT JOIN "idr4_psf"."idr4_psf_r" as r ON tap.ID=r.ID 
            LEFT JOIN "idr4_psf"."idr4_psf_j0660" as j660 ON tap.ID=j660.ID 
            LEFT JOIN "idr4_psf"."idr4_psf_i" as i ON tap.ID=i.ID 
            LEFT JOIN "idr4_psf"."idr4_psf_j0861" as j861 ON tap.ID=j861.ID 
            LEFT JOIN "idr4_psf"."idr4_psf_z" as z ON tap.ID=z.ID"""
```

```python
results = conn.query(Query, df)    
```

```python
results 
```

```python
# Define the value to replace empty values
replacement_value = 99

# Replace empty values with 99 for all columns
for column_name in results.colnames:
    results[column_name] = [value if value != '' and value is not None else replacement_value for value in results[column_name]]
results
```

```python
def create_mask(ra):
    """
    Create a mask for a NumPy array to filter elements greater than a specified threshold.

    Parameters:
        data (numpy.ndarray): The input data array.
        threshold (float): The threshold value for filtering.

    Returns:
        numpy.ndarray: A boolean mask for elements greater than the threshold.
    """
    mask = (ra > 30) & (ra < 70)
    return mask
```

```python
table = results[create_mask(results["RA_r"])]
# Rename columns
table.rename_column('RA_r', 'RA')
table.rename_column('DEC_r', 'DEC')
```
